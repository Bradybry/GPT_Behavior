````python
import ast
def parse_and_run_code(response):
    code = get_code_from_response(response)
    if code:
        try:
            ast.parse(code)
            exec(code)
            return 1
        except Exception:
            return 0
    else:
        return 0

def get_code_from_response(response):
    start = response.find("```python")
    end = response.find("```", start + len("```python"))
    if start != -1 and end != -1:
        return response[start + len("```python"):end]
    else:
        return response
````

````python
# Define the function to parse and run the code
def parse_and_run_code(response):
    code = get_code_from_response(response)
    if code:
        try:
            ast.parse(code)
            exec(code)
            return 1
        except Exception:
            return 0
    else:
        return 0

# Apply the parse_and_run_code function to the "answer" column
df['result'] = df['answer'].apply(parse_and_run_code)
````
